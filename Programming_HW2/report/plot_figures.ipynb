```python
import json
import matplotlib.pyplot as plt 
from matplotlib.scale import LogScale
```

## Comparing MLP and CNN Models

```python
with open('../experiment_result/cnn_model.json', 'r') as f:
    cnn_model = json.load(f)
with open('../experiment_result/mlp_model.json', 'r') as f:
    mlp_model = json.load(f)
```

```python
plt.title("Loss Curve of MLP and CNN Models")
plt.xlabel("Epoches")
plt.ylabel("Cross Entropy Loss")
# plt.legend(["MLP model", "CNN model"])

plt.plot(cnn_model['train_loss'], label="CNN Model")
plt.plot(mlp_model['train_loss'], label="MLP Model")
plt.legend()

plt.savefig(fname="Loss_Curve_MLP_CNN.pdf", format="pdf")
plt.show()
```

## Comparing Activation and No-activation models

```python
with open('../experiment_result/cnn_no_act_model.json', 'r') as f:
    cnn_no_act = json.load(f)
with open('../experiment_result/mlp_no_act_model.json', 'r') as f:
    mlp_no_act = json.load(f)
```

```python
fig, axs = plt.subplots(1,2, figsize=(16,6))

# Plot loss change for MLP model
axs[0].set_title("MLP Model with and without Activation")
axs[0].set_ylabel("Loss")
axs[0].set_xlabel("Epoches")
axs[0].plot(mlp_model["train_loss"], label="With Activation")
axs[0].plot(mlp_no_act["train_loss"], label="No Activation")
axs[0].legend()

# Plot loss change for CNN model
axs[1].set_title("CNN Model with and without Activation")
axs[1].set_ylabel("Loss")
axs[1].set_xlabel("Epoches")
axs[1].plot(cnn_model["train_loss"], label="With Activation")
axs[1].plot(cnn_no_act["train_loss"], label="No Activation")
axs[1].legend()

plt.savefig(fname="Activation_NO_Activation.pdf", format="pdf")
plt.show()
```
