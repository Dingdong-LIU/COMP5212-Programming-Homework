```python
import json
import matplotlib.pyplot as plt
from matplotlib.scale import LogScale
```

## Different Learning Rate

```python
with open('../experiment_results/model_with_lr.json', 'r') as f:
    model_with_lr = json.load(f)
svm_model_history = [x for x in model_with_lr if x['model_type'] == 'svm'][0]
lg_model_history = [x for x in model_with_lr if x['model_type'] == 'logistic regression'][0]
```

```python
for th in svm_model_history['train_history']:
    plt.plot(th['training_history'])
plt.title('SVM Model with Different Step Size')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.yscale(LogScale(axis=0, base=10))
plt.legend([f'lr={x["lr"]}' for x in svm_model_history['train_history']], loc='upper right')
plt.show()

for th in lg_model_history['train_history']:
    plt.plot(th['training_history'])
plt.title('Logistic Regression Model with Different Step Size')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.yscale(LogScale(axis=0, base=10))
plt.legend(
    [f'lr={x["lr"]}' for x in lg_model_history['train_history']], loc='upper right')
plt.show()
```

## Different Value of Momentum

```python
with open("../experiment_results/model_with_momentum.json", 'r') as f:
    model_with_momentum = json.load(f)
svm_momentum = [x for x in model_with_momentum if x['model_type'] == 'svm']
lg_momentum = [x for x in model_with_momentum if x['model_type'] == 'logistic regression']
```

```python
for th in svm_momentum:
    plt.plot(th['train_history'][0]['training_history'])
plt.title('SVM Model with Different Momentum')
plt.ylabel('Accuracy')
plt.yscale(LogScale(axis=0, base=10))
plt.xlabel('Epoch')
plt.legend(
    [f'm={x["train_history"][0]["momentum"]}' for x in svm_momentum], loc='upper right')
plt.show()

for th in lg_momentum:
    plt.plot(th['train_history'][0]['training_history'])
plt.title('Logistic Regression Model with Different Momentum')
plt.ylabel('Accuracy')
plt.yscale(LogScale(axis=0, base=10))
plt.xlabel('Epoch')
plt.legend(
    [f'm={x["train_history"][0]["momentum"]}' for x in lg_momentum], loc='upper right')
plt.show()
```
