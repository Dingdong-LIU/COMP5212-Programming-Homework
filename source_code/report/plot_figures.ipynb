```python
import json
import matplotlib.pyplot as plt
from matplotlib.scale import LogScale
```

## Different Learning Rate

```python
with open('../experiment_results/model_with_lr.json', 'r') as f:
    model_with_lr = json.load(f)
svm_model_history = [x for x in model_with_lr if x['model_type'] == 'svm'][0]
lg_model_history = [x for x in model_with_lr if x['model_type'] == 'logistic regression'][0]
```

```python
fig, axs = plt.subplots(1,2, figsize=(16,6))
for th in svm_model_history['train_history']:
    axs[0].plot(th['training_history'])
axs[0].set_title('SVM Model with Different Step Size')
axs[0].set_ylabel('Accuracy')
axs[0].set_xlabel('Epoch')
axs[0].set_yscale(LogScale(axis=0, base=10))
axs[0].legend([f'lr={x["lr"]}' for x in svm_model_history['train_history']], loc='upper right')
# axs[0].savefig(fname='SVM_Model_with_Different_Step_Size.pdf', format='pdf')
# axs[0].show()

for th in lg_model_history['train_history']:
    axs[1].plot(th['training_history'])
axs[1].set_title('Logistic Regression Model with Different Step Size')
axs[1].set_ylabel('Accuracy')
axs[1].set_xlabel('Epoch')
axs[1].set_yscale(LogScale(axis=0, base=10))
axs[1].legend(
    [f'lr={x["lr"]}' for x in lg_model_history['train_history']], loc='upper right')
plt.savefig(
    fname='Two_Models_with_Different_Step_Size.pdf', format='pdf')
plt.show()
```

## Different Value of Momentum

```python
with open("../experiment_results/model_with_momentum.json", 'r') as f:
    model_with_momentum = json.load(f)
svm_momentum = [x for x in model_with_momentum if x['model_type'] == 'svm']
lg_momentum = [x for x in model_with_momentum if x['model_type'] == 'logistic regression']
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 6))
for th in svm_momentum:
    axs[0].plot(th['train_history'][0]['training_history'])
axs[0].set_title('SVM Model with Different Momentum')
axs[0].set_ylabel('Loss')
axs[0].set_yscale(LogScale(axis=0, base=10))
axs[0].set_xlabel('Epoch')
axs[0].legend(
    [f'momentum={x["train_history"][0]["momentum"]}' for x in svm_momentum], loc='upper right')
# plt.savefig(fname='SVM_Model_with_Different_Momentum.pdf', format='pdf')
# plt.show()

for th in lg_momentum:
    axs[1].plot(th['train_history'][0]['training_history'])
axs[1].set_title('Logistic Regression Model with Different Momentum')
axs[1].set_ylabel('Loss')
axs[1].set_yscale(LogScale(axis=0, base=10))
axs[1].set_xlabel('Epoch')
axs[1].legend(
    [f'momentum={x["train_history"][0]["momentum"]}' for x in lg_momentum], loc='upper right')
plt.savefig(fname='Two_Model_with_Different_Momentum.pdf', format='pdf')
plt.show()
```

## Plot loss function

```python
svm_loss_selected = [x['training_history'] 
                    for x in svm_model_history['train_history'] if x['lr']==0.1][0]
lg_loss_selected = [x['training_history']
                    for x in lg_model_history['train_history'] if x['lr'] == 0.1][0]
plt.plot(svm_loss_selected)
plt.plot(lg_loss_selected)
plt.title('Loss of Two models on Training Set')
plt.legend(["Support Vector Machine", "Logistic Regression"], loc='upper right')
plt.ylabel('Loss')
plt.yscale(LogScale(axis=0, base=10))
plt.xlabel('Epoch')
plt.savefig(fname='Accuracy_with_epoch.pdf', format='pdf')
plt.show()
```
